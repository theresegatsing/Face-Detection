```python
import time
import numpy as np
import h5py
import matplotlib.pyplot as plt
import scipy
from PIL import Image
from scipy import ndimage
import os
from PIL import Image
```

```python
def load_data(data_dir="dataset", img_size=(64, 64)):
    sets = ["train", "test"]
    classes = ["no_face", "face"]

    data = {}

    for s in sets:
        X = []
        y = []

        for label, cls in enumerate(classes):
            class_dir = os.path.join(data_dir, s, cls)

            for file in os.listdir(class_dir):
                img_path = os.path.join(class_dir, file)

                try:
                    img = Image.open(img_path).convert("RGB")
                    img = img.resize(img_size)
                    X.append(np.array(img))
                    y.append(label)
                except:
                    continue  # skip corrupted images

        data[s] = (np.array(X), np.array(y))

    train_x_orig, train_y = data["train"]
    test_x_orig, test_y = data["test"]

    #flattened dataset
    train_x_flatten = train_x_orig.reshape(train_x_orig.shape[0], -1).T  # (12288, 224)
    test_x_flatten = test_x_orig.reshape(test_x_orig.shape[0], -1).T     # (12288, 287)

    #Normalized 
    train_x_orig = train_x_flatten.astype("float32") / 255.0
    test_x_orig = test_x_flatten.astype("float32") / 255.0


    return train_x_orig, train_y, test_x_orig, test_y, np.array(classes)
```

```python
train_x_orig, train_y, test_x_orig, test_y, classes = load_data()
```
